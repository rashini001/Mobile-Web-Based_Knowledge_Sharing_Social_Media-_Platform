```python
import numpy as np
import pandas as pd
```

**Impot Dataset**

```python
data = pd.read_csv('https://raw.githubusercontent.com/hasinduvindana/Education-Recomendation-System/main/datasets/student-scores.csv')
data.head()
```

```python
data = data.drop(['id', 'first_name', 'last_name','email','gender','part_time_job','absence_days','extracurricular_activities','weekly_self_study_hours',], axis=1)
data.head()
```

```python
data.shape
```

```python
data['career_aspiration'].value_counts()
```

**check null data**

```python
data.info()
```

**Get Summery Of data**

```python
data.describe()
```

**Take X values**

```python
x = data.drop(['career_aspiration'],axis=1)
x.head()
```

**Take Y values**

```python
y = data['career_aspiration']
y[0:5]
```

**Featuring Data**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x= pd.DataFrame(scaler.fit_transform(x))
x[0:5]
```

**Taking apart data to test and train**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_test.shape
```

```python
x_train.shape
```

**train data with KKN algorithm**

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=1)
model.fit(x_train, y_train)
```

**check predicted value with test data**

```python
pred = model.predict(x_test)
pred[0:5]
```

```python
y_test[0:5]
```

**check accuracy of model**

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, pred)
accuracy
```

```python
from sklearn.metrics import confusion_matrix
cm = pd.DataFrame(confusion_matrix(y_test, pred))
cm
```

```python
result = pd.DataFrame( data=[y_test.values,pred], index = ['y_test', 'pred'])
result.transpose()
```

**Choose best k values**

```python
correct_sum =  []
for i in range(1,20):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(x_train, y_train)
    pred = model.predict(x_test)
    correct  = np.sum(pred == y_test)
    correct_sum.append(correct)
```

```python
correct_sum
```

```python
result = pd.DataFrame(data=correct_sum)
result.index = result.index + 1
result.T
```

```python
model = KNeighborsClassifier(n_neighbors=13)
model.fit(x_train, y_train)
pred = model.predict(x_test)
```

```python
accuracy_score(y_test, pred)
```
